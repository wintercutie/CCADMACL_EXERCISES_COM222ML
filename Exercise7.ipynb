```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
#reading data
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
dataset = pd.read_csv(url)

dataset.head()
```

```python
client_basket = dataset.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda x: ', '.join(x))
client_basket
```

### 1. Apply one-hot encoding

```python
# If client_basket is a Series, convert it to a list of lists first:
if isinstance(client_basket, pd.Series):
    transactions = []
    for items in client_basket:
        transactions.append(items.split(', ')) # Split the items string
else:
    transactions = client_basket # If already a list of lists, no need to modify


# Applying one-hot encoding
from mlxtend.preprocessing import TransactionEncoder # Import TransactionEncoder
basket_transformer = TransactionEncoder()
basket_data = basket_transformer.fit_transform(transactions)
basket_sets = pd.DataFrame(basket_data, columns=basket_transformer.columns_)

basket_sets.head()
```

2. Use the Apriori Algorithm to find the most frequent sets

```python
frequent_itemsets = apriori(basket_sets, min_support=0.0005, use_colnames=True)
frequent_itemsets
```

3. Display the association rules using the lift metric

```python
# Display the association rules using the lift metric
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules_sorted = rules.sort_values(by='lift', ascending=False)
rules_sorted
```
